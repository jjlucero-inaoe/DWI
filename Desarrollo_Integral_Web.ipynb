```
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar imagen
imagen = cv2.imread('gato.jpg')

# Convertir a float32 y normalizar a rango [0,1]
imagen_norm = imagen.astype(np.float32) / 255.0

# Aplicar la transformaci√≥n: (x/255)^(1/3) * 255
imagen_modificada = np.power(imagen_norm, 1/3.0) * 255.0

# Convertir a uint8
imagen_modificada = imagen_modificada.astype(np.uint8)

# Convertir a RGB para mostrar bien con matplotlib
imagen_rgb = cv2.cvtColor(imagen_modificada, cv2.COLOR_BGR2RGB)

# Mostrar imagen
plt.imshow(imagen_rgb)
plt.axis('off')
plt.title("Contraste no lineal: (x/255)^(1/3) * 255")
plt.show()
```
